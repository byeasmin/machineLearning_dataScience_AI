```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns in your dataset
numerical_columns = ['Total Screen Time', 'Social Networking', 'Reading and Reference',
                     'Other', 'Productivity', 'Health and Fitness', 'Entertainment',
                     'Creativity', 'Yoga']

plt.figure(figsize=(18, 12))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # 3x3 grid for 9 columns
    data = df[column].dropna()
    
    # Plot histogram
    sns.histplot(data, kde=False, color='skyblue', bins=15, label='Data', stat='density')
    
    # Fit and plot the normal distribution curve
    mu, std = stats.norm.fit(data)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)
    plt.plot(x, p, 'r', linewidth=2, label='Normal Distribution')
    
    plt.title(f'Normal Distribution for {column}')
    plt.xlabel(column)
    plt.ylabel('Density')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Numeric columns for QQ plot
numerical_columns = ['Total Screen Time', 'Social Networking', 'Reading and Reference',
                     'Other', 'Productivity', 'Health and Fitness', 'Entertainment',
                     'Creativity', 'Yoga']

plt.figure(figsize=(18, 12))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)  # 3x3 grid for 9 columns
    stats.probplot(df[column].dropna(), dist="norm", plot=plt)
    plt.title(f'QQ Plot for {column}')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of segments in the circle (based on dataset size)
num_segments = len(df)

# Column to visualize
column_to_visualize = 'Total Screen Time'

# Normalize data to 0-1 range for coloring
normalized_data = np.interp(df[column_to_visualize].values, 
                            (df[column_to_visualize].min(), df[column_to_visualize].max()), 
                            (0, 1))

# Angles for each segment in the circle
angles = np.linspace(0, 2 * np.pi, num_segments, endpoint=False)

# Create polar plot
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))

# Plot each segment with color based on the normalized value
for i in range(num_segments):
    ax.bar(angles[i], 1, width=2 * np.pi / num_segments, bottom=0, 
           color=plt.cm.viridis(normalized_data[i]), alpha=0.8)

# Customize plot
ax.set_yticklabels([])  # Remove radial ticks
ax.set_xticklabels([])  # Remove angular ticks
ax.set_title(f"Pixel Layout in Circle Segments for {column_to_visualize}", va='bottom')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns in your dataset
numerical_columns = ['Total Screen Time', 'Social Networking', 'Reading and Reference',
                     'Other', 'Productivity', 'Health and Fitness', 'Entertainment',
                     'Creativity', 'Yoga']

# Compute correlation matrix
corr_matrix = df[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Screen Time Categories')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Numeric columns in your dataset
numerical_columns = ['Total Screen Time', 'Social Networking', 'Reading and Reference',
                     'Other', 'Productivity', 'Health and Fitness', 'Entertainment',
                     'Creativity', 'Yoga']

# Clean data by dropping rows with missing values (if any)
df_cleaned = df.dropna(subset=numerical_columns)

# Generate scatterplot matrix (pairplot)
sns.pairplot(df_cleaned[numerical_columns])
plt.suptitle("Scatterplot Matrix of Screen Time Categories", y=1.02)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Choose three numeric columns
x = df['Social Networking'].values
y = df['Productivity'].values
z = df['Entertainment'].values

# Create a meshgrid for plotting
X, Y = np.meshgrid(np.linspace(x.min(), x.max(), len(x)), 
                   np.linspace(y.min(), y.max(), len(y)))

# Interpolate Z values for demonstration
Z = np.interp(X.flatten(), x, z)
Z = Z.reshape(X.shape)

# Create 3D surface plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none', alpha=0.9)

# Labels and title
ax.set_xlabel('Social Networking')
ax.set_ylabel('Productivity')
ax.set_zlabel('Entertainment')
ax.set_title('3D Landscape Plot of Screen Time Categories')

plt.show()
```
